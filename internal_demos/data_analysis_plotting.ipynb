```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.gridspec as gridspec
import sys
sys.path.append("../")

from statsmodels.graphics.tsaplots import plot_pacf

plt.style.use("seaborn-v0_8")
plt.rcParams["figure.figsize"] = [20, 4]
plt.rcParams.update({"font.family": "serif"})
plt.rcParams["axes.titlesize"] = 16
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10
plt.rcParams["legend.fontsize"] = 12
```

```python
STATIONARY_PATH = "../SYNTHDataset/SYNTHh.csv"
ts_stationary = pd.read_csv(STATIONARY_PATH, index_col=None)
ts_stationary["date"] = pd.to_datetime(ts_stationary["date"])

ADDITIVE_PATH = "../SYNTHDataset/SYNTH_additive.csv"
ts_additive = pd.read_csv(ADDITIVE_PATH, index_col=None)
ts_additive["date"] = pd.to_datetime(ts_additive["date"])
ts_additive["TARGET_diff"] = ts_additive["TARGET"].diff()

ADDITIVE_REVERSAL_PATH = "../SYNTHDataset/SYNTH_additive_reversal.csv"
ts_additive_reversal = pd.read_csv(ADDITIVE_REVERSAL_PATH, index_col=None)
ts_additive_reversal["date"] = pd.to_datetime(ts_additive_reversal["date"])
ts_additive_reversal["TARGET_diff"] = ts_additive_reversal["TARGET"].diff()

MULTIPLICATIVE_PATH = "../SYNTHDataset/SYNTH_multiplicative.csv"
ts_multiplicative = pd.read_csv(MULTIPLICATIVE_PATH, index_col=None)
ts_multiplicative["date"] = pd.to_datetime(ts_multiplicative["date"])
ts_multiplicative["TARGET_diff"] = ts_multiplicative["TARGET"].diff()

MULTIPLICATIVE_REVERSAL_PATH = "../SYNTHDataset/SYNTH_multiplicative_reversal.csv"
ts_multiplicative_reversal = pd.read_csv(MULTIPLICATIVE_REVERSAL_PATH, index_col=None)
ts_multiplicative_reversal["date"] = pd.to_datetime(ts_multiplicative_reversal["date"])
ts_multiplicative_reversal["TARGET_diff"] = ts_multiplicative_reversal["TARGET"].diff()
```

```python
fig = plt.figure(figsize=(12, 9))
gs = gridspec.GridSpec(3, 2, width_ratios=[1, 1])

# Creating subplots
ax0 = plt.subplot(gs[0, :])
ax1 = plt.subplot(gs[1, 0])
ax2 = plt.subplot(gs[1, 1])
ax3 = plt.subplot(gs[2, 0])
ax4 = plt.subplot(gs[2, 1])

# Stationary Series
ax0.plot(ts_stationary["date"], ts_stationary["TARGET"], linewidth=1)
ax0.set_title("Stationary Series")
ax0.set_xlabel("Date")
ax0.set_ylabel("Synthetic Energy Load")
ax0.tick_params(axis="x", rotation=45)

# Additive Trend
ax1.plot(ts_additive["date"], ts_additive["TARGET"], linewidth=1)
ax1.set_title("Series with Additive Trend")
ax1.set_xlabel("Date")
ax1.set_ylabel("Synthetic Energy Load")
ax1.tick_params(axis="x", rotation=45)

# Additive trend and reversals
ax2.plot(
    ts_additive_reversal["date"],
    ts_additive_reversal["TARGET"],
    linewidth=1,
)
ax2.set_title("Series with Additive Trend and Reversals")
ax2.set_xlabel("Date")
ax2.set_ylabel("Synthetic Energy Load")
ax2.tick_params(axis="x", rotation=45)

# Multiplicative trend
ax3.plot(ts_multiplicative["date"], ts_multiplicative["TARGET"], linewidth=1)
ax3.set_title("Series with Multiplicative Trend")
ax3.set_xlabel("Date")
ax3.set_ylabel("Synthetic Energy Load")
ax3.tick_params(axis="x", rotation=45)

# Multiplicative trend and reversals
ax4.plot(
    ts_multiplicative_reversal["date"],
    ts_multiplicative_reversal["TARGET"],
    linewidth=1,
)
ax4.set_title("Series with Multiplicative Trend and Reversals")
ax4.set_xlabel("Date")
ax4.set_ylabel("Synthetic Energy Load")
ax4.tick_params(axis="x", rotation=45)

# Plot and save
plt.tight_layout()
plt.savefig("synth_series_all.pdf", format="pdf")
plt.show()
```

Below we plot the wind time series:

```python
PATH_WIND = "../WINDataset/DEWINDh_small.csv"
df_wind = pd.read_csv(PATH_WIND)
df_wind["date"] = pd.to_datetime(df_wind["date"])
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 9))  # 3 rows, 2 columns

axs[0, 0].plot(
    df_wind.iloc[:24]["date"], df_wind.iloc[:24]["TARGET"], linewidth=2
)
axs[0, 0].set_title("German Wind Generation Capacity (Day)")
axs[0, 0].set_xlabel("Date")
axs[0, 0].set_ylabel("% of Capacity Used")
axs[0, 0].tick_params(axis="x", rotation=45)

axs[1, 0].plot(
    df_wind.iloc[:168]["date"], df_wind.iloc[:168]["TARGET"], linewidth=2
)
axs[1, 0].set_title("German Wind Generation Capacity (Week)")
axs[1, 0].set_xlabel("Date")
axs[1, 0].set_ylabel("% of Capacity Used")
axs[1, 0].tick_params(axis="x", rotation=45)

axs[2, 0].plot(
    df_wind.iloc[:720]["date"], df_wind.iloc[:720]["TARGET"], linewidth=2
)
axs[2, 0].set_title("German Wind Generation Capacity (Month)")
axs[2, 0].set_xlabel("Date")
axs[2, 0].set_ylabel("% of Capacity Used")
axs[2, 0].tick_params(axis="x", rotation=45)

plot_pacf(df_wind.iloc[:24]["TARGET"], ax=axs[0, 1], lags = 11, method="ywm")
axs[0, 1].set_title("PACF: Day")
axs[0, 1].set_xlabel("Lags")
axs[0, 1].set_ylabel("Partial Autocorrelation")

plot_pacf(
    df_wind.iloc[:168]["TARGET"], ax=axs[1, 1], lags=168 // 2 -1, method="ywm"
)
axs[1, 1].set_title("PACF: Week")
axs[1, 1].set_xlabel("Lags")
axs[1, 1].set_ylabel("Partial Autocorrelation")


plot_pacf(
    df_wind.iloc[:720]["TARGET"],
    ax=axs[2, 1],
    lags = 720 // 2 - 1,
    method="ywm",
)
axs[2, 1].set_title("PACF: Month")
axs[2, 1].set_xlabel("Lags")
axs[2, 1].set_ylabel("Partial Autocorrelation")

plt.tight_layout()
plt.savefig("wind_pacf.pdf", format="pdf")
plt.show()
```
